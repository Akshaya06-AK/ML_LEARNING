```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = [
    'PH',
    'HARDNESS',
    'SOLIDS',
    'CHLORAMINES',
  'SULFATE',
  'CONDUCTIVITY',
  'ORGANIC_CARBON',
  'TRIHALOMETHANES',
  'TURBIDITY',
  'POTABILITY'
]

for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))

print(df.head())
```

```python
x=pd.DataFrame(df,columns=['PH','HARDNESS','SOLIDS','CHLORAMINES','SULFATE','CONDUCTIVITY','TRIHALOMETHANES','TURBIDITY'])
y=df['POTABILITY'].values
```

```python
from   sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
x_train['PH'] = x_train['PH'].replace('N/A', np.nan).astype(float)
x_test['PH'] = x_test['PH'].replace('N/A', np.nan).astype(float)

imputer = SimpleImputer(strategy='mean')
x_train['PH'] = imputer.fit_transform(x_train[['PH']])
x_test['PH'] = imputer.transform(x_test[['PH']])
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy score:",metrics.accuracy_score(y_test,y_pred))
print("Prediction of TARGET:",clf.predict([[23, 78, 29,45,30,35,40,45,30]]))
```

```python
from sklearn.metrics import classification_report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
