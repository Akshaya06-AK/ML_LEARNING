```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = [
    'POLICY_NUMBER',
    'POLICY_STATE',
    'POLICY_ANNUAL_PREMIUM',
    'UMBRELLA_LIMIT',
    'INCIDENT_DATE',
    'INCIDENT_TYPE',
    'COLLISION_TYPE',
    'INCIDENT_SEVERITY',
    'AUTHORITIES_CONTACTED',
    'INCIDENT_STATE',
    'INCIDENT_CITY',
    'INCIDENT_LOCATION',
    'NUMBER_OF_VEHICLES_INVOLVED',
    'PROPERTY_DAMAGE',
    'BODILY_INJURIES',
    'WITNESSES',
    'POLICE_REPORT_AVAILABLE',
    'TOTAL_CLAIM_AMOUNT',
    'INJURY_CLAIM',
    'PROPERTY_CLAIM',
    'VEHICLE_CLAIM',
    'AUTO_MAKE',
    'AUTO_MODEL',
    'AUTO_YEAR',
    'FRAUD_REPORTED',
]

for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))

print(df.head())
```

```python
x=pd.DataFrame(df,columns=[ 'POLICY_NUMBER','POLICY_STATE','POLICY_ANNUAL_PREMIUM','UMBRELLA_LIMIT','INCIDENT_DATE','INCIDENT_TYPE','COLLISION_TYPE',
    'INCIDENT_SEVERITY','AUTHORITIES_CONTACTED','INCIDENT_STATE','INCIDENT_CITY','INCIDENT_LOCATION','NUMBER_OF_VEHICLES_INVOLVED','PROPERTY_DAMAGE',
    'BODILY_INJURIES','WITNESSES','POLICE_REPORT_AVAILABLE','TOTAL_CLAIM_AMOUNT','INJURY_CLAIM','PROPERTY_CLAIM','VEHICLE_CLAIM','AUTO_MAKE','AUTO_MODEL',
    'AUTO_YEAR','FRAUD_REPORTED'])
y=df['FRAUD_REPORTED'].values
```

```python
from   sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
x_train['POLICY_NUMBER'] = x_train['POLICY_NUMBER'].replace('N/A', np.nan).astype(float)
x_test['POLICY_NUMBER'] = x_test['POLICY_NUMBER'].replace('N/A', np.nan).astype(float)

imputer = SimpleImputer(strategy='mean')
x_train['POLICY_NUMBER'] = imputer.fit_transform(x_train[['POLICY_NUMBER']])
x_test['POLICY_NUMBER'] = imputer.transform(x_test[['POLICY_NUMBER']])
```

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Apply one-hot encoding to X_train and X_test
x_train_encoded = pd.get_dummies(x_train, columns=['INCIDENT_STATE', 'INCIDENT_LOCATION', 'INCIDENT_SEVERITY'], drop_first=True)
x_test_encoded = pd.get_dummies(x_test, columns=['INCIDENT_STATE', 'INCIDENT_LOCATION', 'INCIDENT_SEVERITY'], drop_first=True)

# Ensure columns are aligned after one-hot encoding, especially if some categories only appear in train or test set
# This step is crucial to prevent errors when X_test might have fewer columns than X_train after get_dummies
missing_cols_train = set(x_test_encoded.columns) - set(x_train_encoded.columns)
for c in missing_cols_train:
    x_train_encoded[c] = 0
missing_cols_test = set(x_train_encoded.columns) - set(x_test_encoded.columns)
for c in missing_cols_test:
    x_test_encoded[c] = 0

x_test_encoded = x_test_encoded[x_train_encoded.columns]

model= DecisionTreeClassifier(criterion = 'entropy')
model.fit(x_train_encoded,y_train)
```
