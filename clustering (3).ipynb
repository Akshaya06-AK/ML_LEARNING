```python
from sklearn.neighbors import NearestNeighbors

# 'q' contains your numerical data from the autos dataset
# Let's set min_samples based on the number of features in q
n_features = q.shape[1]
min_samples = 2 * n_features # A common heuristic
print(f"Suggested min_samples: {min_samples}")

# Calculate the distance to the k-th (min_samples) nearest neighbor for each point
neigh = NearestNeighbors(n_neighbors=min_samples)
neigh.fit(q)
distances, indices = neigh.kneighbors(q)

# Sort the distances in ascending order
distances = np.sort(distances[:, min_samples-1], axis=0)

# Plot the K-distance graph
plt.figure(figsize=(12, 6))
plt.plot(distances)
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{min_samples}-th Nearest Neighbor Distance (Epsilon)')
plt.title('K-distance Graph for DBSCAN Epsilon Estimation')
plt.grid(True)
plt.show()

print("Look for the 'elbow' point in the graph. The y-value at this elbow is a good candidate for your `eps`.")
```

```python
# Replace 'your_chosen_eps' with the value you found from the K-distance plot
your_chosen_eps = 1500 # This is an arbitrary placeholder, you need to pick it from the plot

# Re-initialize DBSCAN with the new parameters
db_tuned = DBSCAN(eps=your_chosen_eps, min_samples=min_samples, metric='euclidean')

# Fit the model and get the new labels
model_tuned = db_tuned.fit(q)
labels_tuned = model_tuned.labels_

# Add the cluster labels to your original DataFrame for easier analysis
df['DBSCAN_Cluster'] = labels_tuned

print(f"Number of clusters found: {len(set(labels_tuned)) - (1 if -1 in labels_tuned else 0)}")
print(f"Number of noise points: {np.sum(labels_tuned == -1)}")
print("New labels:")
display(df['DBSCAN_Cluster'].value_counts())
```

```python
plt.figure(figsize=(10, 8))

# Get unique labels (clusters)
unique_labels = set(labels_tuned)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

# Extract relevant columns for plotting from the original df
x_plot = df['horsepower'].values
y_plot = df['price'].values

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'
        marker = 'x'
        # Continue to the next iteration if no noise points to plot, to avoid error.
        if np.sum(labels_tuned == k) == 0: continue
    else:
        marker = 'o'

    # Select points belonging to the current cluster k
    class_member_mask = (labels_tuned == k)

    xy = q[class_member_mask]
    # Ensure we use the correct original columns for plotting, not just arbitrary q columns
    plt.scatter(x_plot[class_member_mask],
                y_plot[class_member_mask],
                c=[col], marker=marker, s=50, label=f'Cluster {k}')

plt.title(f'DBSCAN Clustering (eps={your_chosen_eps}, min_samples={min_samples})')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv('/content/drive/MyDrive/akshaya /Mall_Customers.csv')

df.head()

df.tail()

df.info()

df.describe()

df.shape

df.isnull().sum()

sns.pairplot(df)
plt.show()

sns.scatterplot(
    x='Age',
    y='Annual_Income_(k$)',
    data=df
)

plt.show()

q = df.iloc[:,3:].values

q

import scipy.cluster.hierarchy as sch

plt.figure(figsize=(12,6))

dendrogram = sch.dendrogram(
    sch.linkage(
        q,
        method='ward'
    )
)

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')

plt.show()

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(
    n_clusters=5,
    metric='euclidean',
    linkage='ward'
)

label = hc.fit_predict(q)

label

np.unique(
    label,
    return_counts=True
)

plt.figure(figsize=(10,6))

plt.scatter(
    q[label==0,0],
    q[label==0,1],
    s=80,
    label='Cluster 1'
)

plt.scatter(
    q[label==1,0],
    q[label==1,1],
    s=80,
    label='Cluster 2'
)

plt.scatter(
    q[label==2,0],
    q[label==2,1],
    s=80,
    label='Cluster 3'
)

plt.scatter(
    q[label==3,0],
    q[label==3,1],
    s=80,
    label='Cluster 4'
)

plt.scatter(
    q[label==4,0],
    q[label==4,1],
    s=80,
    label='Cluster 5'
)

plt.title('Hierarchical Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()

plt.show()

from sklearn.metrics import silhouette_score

score = silhouette_score(q, label)

print("Silhouette Score:", score)
```

```python
transactions=[]
for i in range(0,len(df)):
  transactions.append([str(df.values[i,j]) for j in range(0,20)if str(df.values[i,j])!='nan'])
```

```python
print(transactions)
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
results=list(rules)
results
```

```python
import pandas as pd

data = []

for rule in results:

    support = rule.support

    for stat in rule.ordered_statistics:

        lhs = ', '.join(list(stat.items_base))
        rhs = ', '.join(list(stat.items_add))
        confidence = stat.confidence
        lift = stat.lift

        data.append([lhs, rhs, support, confidence, lift])

df_rules = pd.DataFrame(
    data,
    columns=['LHS', 'RHS', 'Support', 'Confidence', 'Lift']
)

print(df_rules)
```
